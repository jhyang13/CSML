```python
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure(figsize=(10,20))#figure size x & y 
n=5 # number of your data sets
ax = [fig.add_subplot(n,1,i+1) for i in range(n)]


fig.subplots_adjust(hspace=0)
x=np.linspace(0,1) # no need if the data is [x,y] 
ax[0].plot(x, x ** 2) # data1
ax[0].locator_params(axis='y', nbins=2)
for i in range(1,n):
  ax[i].spines['top'].set_visible(False)
  ax[i].plot(x, x ** 2) # data 2 ,3,4, ....
  ax[i].locator_params(axis='y', nbins=2)
  

plt.xlabel('Temperature')
plt.ylabel('') # ylabel
plt.show()
```
